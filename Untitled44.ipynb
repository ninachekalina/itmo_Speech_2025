```python
from typing import Optional
import torch
from torch import nn
from torchaudio import functional as F

class LogMelFilterBanks(nn.Module):
    def __init__(
            self,
            n_fft: int = 400,
            samplerate: int = 16000,
            hop_length: int = 160,
            n_mels: int = 80,
            pad_mode: str = 'reflect',
            power: float = 2.0,
            normalize_stft: bool = False,
            onesided: bool = True,
            center: bool = True,
            return_complex: bool = True,
            f_min_hz: float = 0.0,
            f_max_hz: Optional[float] = None,
            norm_mel: Optional[str] = None,
            mel_scale: str = 'htk'
        ):
        super(LogMelFilterBanks, self).__init__()
        # Параметры основные
        self.n_fft = n_fft
        self.samplerate = samplerate
        self.hop_length = hop_length
        self.n_mels = n_mels
        self.f_min_hz = f_min_hz
        self.f_max_hz = f_max_hz if f_max_hz is not None else samplerate / 2
        self.norm_mel = norm_mel
        self.mel_scale = mel_scale
        self.window = torch.hann_window(n_fft)

        # Инициализация mel filter banks
        self.mel_fbanks = self._init_melscale_fbanks()

    def _init_melscale_fbanks(self):
        n_freqs = self.n_fft // 2 + 1  # Количество частотных бинов
        return F.melscale_fbanks(
            n_freqs=n_freqs,
            f_min=self.f_min_hz,
            f_max=self.f_max_hz,
            n_mels=self.n_mels,
            sample_rate=self.samplerate,
            norm=self.norm_mel,
            mel_scale=self.mel_scale
        )

    def spectrogram(self, x):
        # Вычисление the Short-Time Fourier Transform (STFT)
        stft_result = torch.stft(
            x,
            n_fft=self.n_fft,
            hop_length=self.hop_length,
            window=self.window,
            center=True,
            pad_mode='reflect',
            normalized=False,
            onesided=True,
            return_complex=True
        )

        # вычисление power spectrum
        power_spec = (stft_result.real ** 2 + stft_result.imag ** 2)
        return power_spec

    def forward(self, x):
        """
        Args:
            x (Torch.Tensor): Tensor of audio of dimension (batch, time), audio signal
        Returns:
            Torch.Tensor: Tensor of log mel filterbanks of dimension (batch, n_mels, n_frames),
                where n_frames is a function of the window_length, hop_length and length of audio
        """
        # вычисляем power spectrogram
        power_spec = self.spectrogram(x)
        print("power_spec shape:", power_spec.shape)
        print("mel_fbanks shape:", self.mel_fbanks.shape)


        #  mel filter banks
        mel_energies = torch.matmul(power_spec.transpose(1, 2), self.mel_fbanks)
        print("mel_energies shape:", mel_energies.shape)


        # Convert to log scale
        #log_mel_energies = torch.log1p(mel_energies)  # log(1 + x) for numerical stability
        log_mel_energies = torch.log(mel_energies + 1e-6)

        return log_mel_energies
```

```python
import torchaudio
import matplotlib.pyplot as plt

# Загрузите аудиофайл
wav_path = '/content/Wav_868kb.wav'  #  путь к  аудиофайлу
signal, sr = torchaudio.load(wav_path)

# проверка что сигнал моно -1 канал
if signal.shape[0] > 1:
    signal = signal.mean(dim=0, keepdim=True)

# Инициализация LogMelFilterBanks
logmel_filterbanks = LogMelFilterBanks(samplerate=sr, hop_length=160, n_mels=80)

# Вычисление логарифм. мел-спектр с использованием  реализации
logmel_spec = logmel_filterbanks(signal)
# Вывод shape -отладкa
print(f"Shape of logmel_spec (custom implementation): {logmel_spec.shape}")



# мел-спектр с использованием torchaudio
melspec = torchaudio.transforms.MelSpectrogram(
    sample_rate=sr,
    n_fft=400,
    hop_length=160,
    n_mels=80
)(signal)
print(f"Shape of melspec (torchaudio): {torch.log(melspec + 1e-6).shape}")

# Визуализация логарифмический мел-спектр
plt.figure(figsize=(10, 4))
plt.imshow(logmel_spec[0].detach().numpy(), aspect='auto', origin='lower')
plt.title('Log Mel Spectrogram (Custom Implementation)')
plt.xlabel('Frames')
plt.ylabel('Mel Bands')
plt.colorbar(format='%+2.0f dB')
plt.show()

# Визуализация мел-спектр с помощью torchaudio
plt.figure(figsize=(10, 4))
plt.imshow(melspec[0].detach().numpy(), aspect='auto', origin='lower')
plt.title('Native Mel Spectrogram (torchaudio)')
plt.xlabel('Frames')
plt.ylabel('Mel Bands')
plt.colorbar(format='%+2.0f dB')
plt.show()

logmel_spec = logmel_spec.transpose(1, 2)  #к форме (batch, n_mels, n_frames)

# Проверка shape -размерности
assert logmel_spec.shape == torch.log(melspec + 1e-6).shape, "Ошибка: Размерности не совпадают!"

# Сравнение - разница между двумя матрицами
diff = torch.abs(logmel_spec - torch.log(melspec + 1e-6))
print(f"Максимальная разница: {diff.max().item()}")
print(f"Средняя разница: {diff.mean().item()}")

# Отображаем первые несколько значений
print("Первые 5 значений из logmel_spec (custom):", logmel_spec.flatten()[:5])
print("Первые 5 значений из melspec (torchaudio):", torch.log(melspec + 1e-6).flatten()[:5])


# Проверка на точность  с  погрешностью
assert torch.allclose(logmel_spec, torch.log(melspec + 1e-6), atol=1e-4), "Ошибка: Результаты не совпадают с допустимой погрешностью"
```
